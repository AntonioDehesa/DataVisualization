```python
# imports
import pandas as pd
import numpy as np
```

```python
carnegie = pd.read_csv("./resources/carnegie2021.csv", encoding="ISO-8859-1")
census = pd.read_csv("./resources/us_census_regions_divisions.csv")
```

```python
carnegie = carnegie.drop(["basic2000", "basic2005", "basic2010", "basic2015" ,"basic2018"], axis=1)
carnegie = carnegie[(carnegie["basic2021"] == 15) | (carnegie["basic2021"] == 16)]
```

```python
carnegie[["City", "State Code"]] = carnegie["location"].str.split(",", expand=True)
carnegie
```

```python
data = pd.merge(carnegie, census, on="State Code", how="left")
```
