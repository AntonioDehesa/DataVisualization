# Part 4

```python
iris = sns.load_dataset("iris")
```

```python
print(iris.head(n=10))
print(iris.info())
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
ax0 = axs[0,0]
ax1 = axs[0,1]
ax2 = axs[1,0]
ax3 = axs[1,1]
ax0.set_title('Sepal Length')
ax1.set_title('Sepal Width')
ax2.set_title('Petal Length')
ax3.set_title('Petal Width')
sns.boxplot(data=iris, x="sepal_length", ax=ax0, hue="species")
sns.boxplot(data=iris, x="sepal_width", ax=ax1, hue="species")
sns.boxplot(data=iris, x="petal_length", ax=ax2, hue="species")
sns.boxplot(data=iris, x="petal_width", ax=ax3, hue="species")
plt.tight_layout()
plt.show()
```
